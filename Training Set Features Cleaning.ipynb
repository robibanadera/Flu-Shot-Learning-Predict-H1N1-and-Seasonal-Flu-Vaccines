```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'training_set_features.csv')
df = df.drop(columns=['employment_industry', 'employment_occupation', 'hhs_geo_region'], axis=1)
df.head(5)
```

```python
float_cols = df.columns[df.dtypes == float]
str_cols = df.columns[df.dtypes == object]

df[float_cols] = df[float_cols].fillna(df[float_cols].mode().iloc[0])
df[str_cols] = df[str_cols].fillna('Missing Data')
display(df)
```
